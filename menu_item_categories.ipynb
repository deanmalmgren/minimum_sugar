# Purpose
In this notebook I work out a way to separate the menu items into various categories.

```python
import json
import minimum_sugar

# Load credential data from file
with open("credentials.json", "r") as f:
    credentials = json.load(f)
    
# Load menu data from file
with open("restaurant_menu_data.json", "r") as f:
    restaurant_menu_data = json.load(f)
```

```python
param = "item_name"
restaurant_menu_item_names = {}

for menu_data in restaurant_menu_data:
    menu_item_names = minimum_sugar.extract_variable(menu_data, param)
    menu_item_names.sort()
    restaurant_menu_item_names[menu_data["name"]] = menu_item_names
```
